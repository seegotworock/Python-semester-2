```python
import numpy as np

def isDiagonallyDominant(x):
    x = np.array(x) ##ubah list jadi -> array
    x = np.abs(x) #dijadika absoolute
    diag = np.diag(x) #ambil diagonal [10,6,6]
    # axis 1 -> untuk baris
    of_diag = np.sum(x, axis = 1) -  diag  #[9 2 4]

    return np.all(diag>of_diag)

def gauss_siedel(A, b, threshold= 0.01, it= 15):
    if not isDiagonallyDominant(A):
        print("Is diagonally Dominant")
        return
    
    A = np.array(A)
    b = np.array(b)

    diag = np.diag(A) #diagonal buat bagi
    A = -A
    np.fill_diagonal(A, 0)

    x_old = np.zeros(A.shape[0]) #[0 0 0]

    for i in range(it):
        x_new = np.array(x_old) #copy value x lama -> x baru

        # b -> [8 5 10]
        for idx, row in enumerate(A):
            x_new[idx] = (b[idx] + np.dot(x_new, row)) / diag[idx]

        print(f'Iter - {i} {x_new}')

        dx = np.sqrt(np.dot(x_new - x_old, x_new - x_old))
        if dx < threshold:
            print('converge')
            return x_new
        
        x_old = x_new 

    print('Not Cnnverged')


```

```python
Xs = [
    [10,4, 5],
    [1,6,1],
    [3,1,6]
]

Ys = [8, 5, 10]

gauss_siedel(Xs, Ys)
```
