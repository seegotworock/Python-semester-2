```python
import numpy as np
```

```python
def f(x):
    return x**2 - 4

def g(x):
    return 2*x  
```

```python
print(f(4)) #14
print(np.sign(f(4)))
print(f(0)) # - 2
print(np.sign(f(0)))
```

```python
def bisection(a, b, tol = 0.001):
    if(np.sign(f(a))) == np.sign(f(b)):
        print('interval is not valid')
        return
    
    c = (a+b)/2
    print("Iteration")
    print(f'a: {a}, b: {b}, c: {c}')
    print(f"f(c) = ", f(c))


    if f(c)==0:
        print("Root is ", c)
        return
    
    if np.abs(f(c)) <= tol:
        print('Hit Tolarance. Root is ', c)
        return
    
    if np.sign(f(c)) == np.sign(f(a)):
        bisection(c, b)
    elif(np.sign(f(c))) == np.sign(f(c)):
        bisection(a,c)
    
```

```python
bisection(0, 4)
```

```python
#Newton_Raphson
def  newton_raphson(x0, tol = 0.001, iter = 100):
    # x1 = x0 - f(x)/g(x)
    x = x0
    for i in range(iter):
        if g(x) == 0:
            print("Derivative is zero")
            return
        
        x_new = x - f(x)/g(x)
        print(f'Iteration {i+1}: x = {x_new} y = {f(x_new)}')

        if abs(f(x_new)) <= tol:
            print("Root is ", x_new)
            return
        
        x= x_new
    
    print("Did not Converge")
```

```python
newton_raphson(4)
```
