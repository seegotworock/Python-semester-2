```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# matplotlib introduction
x = [1,2,3,4,5,6,7,8]
y = [2,5,1,7,9,3,6,10]

plt.plot(x, y, "go", label ="data points")
plt.plot(x, y, "-r", label ="data line")
plt.title("Basic plotting")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# least square regression
x = [1,2,3,4,5,6,7,8]
y = [2,5,1,7,9,3,6,10]

#polyfit
x = np.array(x) #list -> array
y = np.array(y)
                 #      m(slope)      b(intercept)
np.polyfit(x,y,1) # array([0.82142857, 1.67857143])
```

```python
#normal method (quizz)

n = len(x)
X = np.sum(x)
Y = np.sum(y)
XY = x @ y # @ -> perkalian matrix
X2 = x @ x

m = (n*XY - X*Y) / (n*X2 - X**2)
b = (Y- m*X) / n
print(m,b)
```

```python
# Prepapre matrix A and y
ones_arr = np.ones(x.shape)
A = np.vstack ((x, ones_arr)).T # T for transpose and vertical stack for this columb
Y = y[:, np.newaxis]
```

```python
# Direct method (quiz)
alpha = np.linalg.inv(A.T @ A) @ A.T @ Y
alpha
```

```python
#pseudoinverse method
alpha = np.linalg.pinv(A) @ Y
alpha
```

```python
#plotting  rumsu linear = (y=mx+b)
plt.plot(x, y,'ok')
plt.plot(x,alpha[0]*x + alpha[1])
plt.title('least square regression')
plt.show()
```

```python
# polynomial regression
#plot for 6 degrees of fit
for i in range(1, n-1):
    plt.subplot(2, 3, i)
    plt.plot(x,y,'ro')

    #cari coeffficient
    coeffs = np.polyfit(x, y, i)
    y_fit = np.polyval(coeffs, x)
    plt.plot(x, y_fit)
    plt.title(f'Polynomial degree {i}')

plt.tight_layout()
plt.show()
```

```python
#Linear Interpolation
x = [1,2,3,4,5,6,7,8]
y = [2,5,1,7,9,3,6,10]

#to get 100 evenly spaced points from the lowest value in x to the highest value in x
x_new = np.linspace(min(x), max(x), 100)
y_new = np.interp(x_new, x, y)

plt.plot(x,y,'ok')
plt.plot(x_new, y_new, 'r-')
plt.show()
```

```python
#Cubic interpolation
#import Library

from scipy.interpolate import CubicSpline

f = CubicSpline(x, y)

x_new = np.linspace(min(x), max(x), 100)
y_new = f(x_new)

plt.plot(x,y,'ok')
plt.plot(x_new, y_new, 'r-')
plt.show()
```
