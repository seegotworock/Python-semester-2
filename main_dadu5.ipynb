```python
start = 0 # mulai range
end = np.pi # akhir range
n = 1000 # mau bagi area jadi beberapa section
base =(end - start)/n # alas tiap persegi panjang/section

x = np.linspace(start, end, n+1)
x_mid = (x[1:]+x[:-1])/2

y_left = f(x[:-1]) # target sebelah kiri setiap section
y_right = f(x[1:]) # target sebelah kanan setiap section
y_mid = f(x_mid)


left_riemann = np.sum(y_left)* base #jumlhin seua tingii, baru dikali alas
right_riemann = np.sum(y_right)* base
mid_riemann = np.sum(y_mid)* base

area = (y_left + y_right)/2 * base # area per trapezoid
trapezoid_riemann = np.sum(area) # total all trapezoid area

print("REal answer: 2")
print(f"left riemann: {left_riemann}")
print(f"right riemann: {right_riemann}" )
print(f"mid riemann: {mid_riemann}")
print(f"Trapezoid riemann: {trapezoid_riemann}")
```
